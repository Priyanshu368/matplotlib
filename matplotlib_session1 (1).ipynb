```python
#labels title
#use of grid ( axis ='x'/'y'/both,color ='color',linestyle ='linestyle',linewidth=20) 

plt.title('Rohit Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
batsman =pd.read_csv('sharma-kohli.csv')
batsman.head()


plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.grid()
```

```python
#colors(hex) and line (width and style)and marker (size)
#linestyle ='solid'/'dotted'/'dashed'/'dashdot'

#marker ='D'/'<'/'>'/'o'
#markersize =  integer value   ----tells about size of marker
#plt.legend(loc ='best'/'upper right'/'upper left'/''right'/'center left'/'center'/'right'/') ----when we have to lable  to tells which graph is for which player in  this case label ='kohli', label ='rohit'

batsman =pd.read_csv('sharma-kohli.csv')
plt.plot(batsman['index'],batsman['V Kohli'],markeredgecolor ='green',markerfacecolor ='r',linestyle='dashed',linewidth =3,marker='o',markersize =10,label ='kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color ='black',linestyle ='dotted',linewidth =2,marker ='D',markersize =5,label ='rohit')
font1={'family':'serif','color':'blue','size':15} 
font2={'family':'serif','color':'#ffdf4f','size':20}
plt.title('Rohit Vs Virat Kohli Career Comparison',fontdict =font1)
plt.xlabel('Season',fontdict=font2)
plt.ylabel('Runs Scored',fontdict =font2)
plt.legend(loc ='best')   
plt.grid()
```

```python
#color and label--
df =pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])# graph overpass each other
```

```python
plt.bar(np.arange(df.shape[0])-0.3,df['2015'],width =0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width =0.2) #multiple bar charts
plt.bar(np.arange(df.shape[0])+0.3,df['2017'],width =0.2)
```

```python
##xticks --function
plt.bar(np.arange(df.shape[0])-0.3,df['2015'],width =0.2,color ='red')
plt.bar(np.arange(df.shape[0]),df['2016'],width =0.2,color ='blue')
plt.bar(np.arange(df.shape[0])+0.3,df['2017'],width =0.2,color ='yellow')
plt.xticks(np.arange(df.shape[0]),df['batsman']) # plt.xticks(existing name ,new name)
plt.show()
```

```python
#when we have large categorical name such as  example == 'OrangeOrangeOrangeOrangeOrange'
#simple bar chart analysis 
children =[10,20,30,50,76]
colors =['RedRedRedRedRed','BlueBlueBlueBlueBlueBlueBlue','WhiteWhiteWhiteWhiteWhite','OrangeOrangeOrangeOrangeOrange','Yellow']
plt.bar(colors,children,color ='yellow')
```

```python
#to overcome above problem we can use xticks(rotation ='verical ') 
children =[10,20,30,50,76]
colors =['RedRedRedRedRed','BlueBlueBlueBlueBlueBlueBlue','WhiteWhiteWhiteWhiteWhite','OrangeOrangeOrangeOrangeOrange','Yellow']
plt.bar(colors,children,color ='yellow')
plt.xticks(rotation ='vertical')
```

```python
#stacked bar charts
plt.bar(df['batsman'],df['2017'],color ='blue',label ='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],color ='green',label ='2016')
plt.bar(df['batsman'],df['2015'],bottom=df['2016']+df['2017'],color='yellow',label ='2015')
plt.legend()
plt.show()
```

```python
#simple data --
data =[23,4,5,200,47]
sub=['ENGLISH','SCIENCE','MATHS','SST','HINDI']
plt.pie(data,labels=sub)
plt.show()
```

```python
#work on some data ==
df =pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels =df['batsman'],autopct ='%0.1f%%',colors =['blue','green','yellow','pink','cyan','brown'])


plt.show()
```

```python
#explode ,shadow
plt.pie(df['batsman_runs'],labels =df['batsman'],autopct ='%0.1f%%',explode=[0.1,0,0,0,00,0.1],shadow =True)
plt.show()
```
